```python
## Hypergeometric
[M, n, N] = [20, 7, 12]
k = np.arange(0, n + 1)
hypergeom = stats.hypergeom.pmf(k, M, n, N)
print(hypergeom)
plt.plot(k, hypergeom, 'o-')
plt.title('Hyper Geometric:  M:{} n: {} N: {}' .format(M, n, N))
plt.xlabel('Başarı sayısı')
plt.ylabel('Başarı ihtimali')
plt.show()
```
